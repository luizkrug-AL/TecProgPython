```python
# Element-wise e Broadcasting

salarios = np.array([3000, 4000, 5000, 6000])

# Broadcasting: multiplicar cada salário por 1.1 (aumento de 10%)
salarios_aumentados = salarios * 1.1
print(salarios_aumentados)

# Element-wise: somar um array de bônus a cada salário
bonus = np.array([500, 600, 700, 800])
salarios_com_bonus = salarios + bonus
print(salarios_com_bonus)
```

```python
# Funções Matemáticas e Estatísticas
faturamento_diario = np.array([1000, 1500, 2000, 2500, 3000])

# Total da semana
print(f"Total da semana: R$ {np.sum(faturamento_diario):,.2f}")
# media diária
print(f"Média diária: R$ {np.mean(faturamento_diario):,.2f}")
# Desvio padrão da semana
print(f"Desvio padrão: R$ {np.std(faturamento_diario):,.2f}")
# Pior dia
print(f"Pior dia: R$ {np.min(faturamento_diario):,.2f}")
# Melhor dia
print(f"Melhor dia: R$ {np.max(faturamento_diario):,.2f}")

print(f"Índice do melhor dia: {np.argmax(faturamento_diario)}")
print(f"Índice do pior dia: {np.argmin(faturamento_diario)}")
```

### Exercício 3 — Médio: Operações e funções estatísticas

**Enunciado:** Uma padaria registrou suas vendas (em R$) de segunda a sábado: `[450.00, 320.50, 510.75, 489.00, 605.25, 780.50]`. Com esse array:

1. Calcule o total faturado na semana.
2. Calcule a média diária de vendas.
3. Identifique o menor e o maior faturamento.
4. Aplique um desconto de 5% em todos os dias (simule uma promoção onde a padaria deu 5% de cashback). Exiba os valores após o desconto.

```python
# Criando o array de vendas de segunda a sábado
vendas = np.array([450.00, 320.50, 510.75, 489.00, 605.25, 780.50])
print(f"Vendas da semana: {vendas}")

# 1. Calcule o total de vendas da semana
total_vendas = np.sum(vendas)
print(f"Total de vendas da semana: R$ {total_vendas:.2f}")

# 2. Calcule a média diária de vendas
media_vendas = np.mean(vendas)
print(f"Média diária de vendas: R$ {media_vendas:.2f}")

# 3. Identifique o menor e o maior faturamento
min_vendas = np.min(vendas)
max_vendas = np.max(vendas)
print(f"O menor faturamento foi: R$ {min_vendas:.2f}")
print(f"O maior faturamento foi: R$ {max_vendas:.2f}")

# 4. Aplique um desconto de 5% em todos os dias (simule uma promoção onde a padaria deu 5% de cashback).
# Exiba os valores após o desconto.
vendas_com_desconto = vendas * 0.95
print(f"Vendas após o desconto de 5%: R$ {vendas_com_desconto.round(2)}")
```

### Exercício 4 — Médio: Arange e Linspace

**Enunciado:**

a) Use `np.arange()` para criar um array com todos os números pares de 0 a 20 (inclusive).  
b) Use `np.linspace()` para criar um array com exatamente 7 valores igualmente espaçados entre 0% e 100% (representando faixas de desconto de uma promoção progressiva).  
c) Exiba ambos os arrays e seus respectivos `shape`.


```python
# a) Use np.arange para criar um array com todos os números pares de 0 a 20 (inclusive).
numeros_pares = np.arange(0, 21, 2)

# b) use np.linspace() para criar um array com exatamente 7 valores igualmente espaçados entre 0% e 100%
# (representando faixas de desconto de uma promoção progressiva)
descontos = np.linspace(0, 100, 7)

# c) Exiba ambos os arrays e seus respectivos shape.
print(f"Números pares de 0 a 20: {numeros_pares}")
print(f"Shape dos números pares: {numeros_pares.shape}")
print(f"Faixas de desconto (em %): {descontos}")
print(f"Shape das faixas de desconto: {descontos.shape}")
```

### Exercício 5 — Desafio: Análise comparativa com arrays

**Enunciado:** Duas equipes de vendas registraram seus resultados mensais ao longo de 6 meses:

- **Equipe A:** `[42000, 38500, 51000, 47300, 55800, 49200]`
- **Equipe B:** `[39000, 41200, 48700, 52000, 44600, 53100]`

Resolva as seguintes questões **sem usar laços `for`** (use apenas operações NumPy):

1. Qual equipe teve o maior faturamento **total** em 6 meses?
2. Calcule a diferença de vendas mês a mês (Equipe A − Equipe B). Em quais meses a Equipe A superou a Equipe B?
3. Calcule a média de cada equipe e o desvio padrão. Qual equipe é mais consistente (menor desvio)?
4. Crie um array `bonus` onde cada equipe recebe 8% do seu próprio faturamento mensal como bônus. Exiba o total de bônus de cada equipe.

```python
Equipe_A = np.array([42000, 38500, 51000, 47300, 55800, 49200])
Equipe_B = np.array([39000, 41200, 48700, 52000, 44600, 53100])

# 1. Qual equipe teve o maior faturamento total em 6 meses?
total_equipe_a = np.sum(Equipe_A)
total_equipe_b = np.sum(Equipe_B)
if total_equipe_a > total_equipe_b:
    print(f"A Equipe A teve o maior faturamento total: R$ {total_equipe_a:,.2f}")
else:
    print(f"A Equipe B teve o maior faturamento total: R$ {total_equipe_b:,.2f}")
print()
# 2. Calcule a diferença de vendas mês a mês (Equipe A - Equipe B). Em quais meses a Equipe A superou a Equipe B?
diferenca_mensal = Equipe_A - Equipe_B
print(f"Diferença de vendas mês a mês (Equipe A - Equipe B): {diferenca_mensal}")
meses_a_supera_b = diferenca_mensal > 0
'print(f"Meses em que a Equipe A superou a Equipe B: {meses_a_supera_b}")'
print(f"Meses em que a Equipe A superou a Equipe B (índices): {np.where(meses_a_supera_b)[0]}")
print()
# 3. Calcule a média de cada equipe e o desvio padrão. Qual equipe é mais consistente (menor desvio)?
media_a = np.mean(Equipe_A)
desvio_a = np.std(Equipe_A)
print(f" Média Equipe A: {media_a:,.2f}, Desvio Padrão A: {desvio_a:,.2f}")
media_b = np.mean(Equipe_B)
desvio_b = np.std(Equipe_B)
print(f" Média Equipe B: {media_b:,.2f}, Desvio Padrão B: {desvio_b:,.2f}")
if desvio_a < desvio_b:
    print("A Equipe A é mais consistente.")
elif desvio_b < desvio_a:
    print("A Equipe B é mais consistente.")
else:
    print("Ambas as equipes têm o mesmo desvio padrão, portanto são igualmente consistentes.")
print()
# 4. Crie um array `bonus` onde cada equipe recebe 8% do seu próprio faturamento mensal como bônus. 
# Exiba o total de bônus de cada equipe.
bonus_a = Equipe_A * 0.08
bonus_b = Equipe_B * 0.08
total_bonus_a = np.sum(bonus_a)
total_bonus_b = np.sum(bonus_b)
print(f"Total de bônus da Equipe A: R$ {total_bonus_a:,.2f}")
print(f"Total de bônus da Equipe B: R$ {total_bonus_b:,.2f}")
```
