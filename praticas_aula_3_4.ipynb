### Exercício 1 — Fácil: Criando e inspecionando matrizes

**Enunciado:** Crie as seguintes matrizes e exiba o `shape`, `ndim` e `size` de cada uma:

a) Uma matriz 3×3 com os valores de 1 a 9 (use `arange` + `reshape`).  
b) Uma matriz de zeros com 4 linhas e 2 colunas.  
c) A matriz identidade de tamanho 3.

````python
import numpy as np

# Uma matriz 3x3 com valore de 1 a 9
matriz_3x3 = np.arange(1, 10)

print(matriz_3x3.reshape(3,3))
print("shape:", matriz_3x3.shape)
print("ndim:", matriz_3x3.ndim)
print("size:", matriz_3x3.size)
print()
# Uma matriz de zeros com 4 linhas e 2 colunas
matriz_zeros = np.zeros((4,2))
print(matriz_zeros)
print("shape:", matriz_zeros.shape)
print("ndim:", matriz_zeros.ndim)
print("size:", matriz_zeros.size)
print()

#A matriz identidade de tamanho 3
matriz_ident = np.eye(3)
print(matriz_ident)
print("shape:", matriz_ident.shape)
print("ndim:", matriz_ident.ndim)
print("size:", matriz_ident.size)
````

vendas_ano = np.array([45, 52, 48, 60, 55, 70, 80, 78, 65, 58, 50, 90])

### Exercício 2 — Fácil: Indexação e fatiamento

**Enunciado:** Dada a matriz de temperaturas diárias de 3 cidades ao longo de 5 dias:

```python
temperaturas = np.array([
    [32, 33, 31, 29, 35],   # Cidade A
    [28, 27, 30, 26, 29],   # Cidade B
    [22, 24, 21, 23, 25]    # Cidade C
])
```

Extraia:
- A temperatura da Cidade B no 3º dia (índice 2).
- Todas as temperaturas da Cidade C.
- As temperaturas de todas as cidades no último dia.
- A submatriz das cidades A e B nos primeiros 3 dias.


````python
temperaturas = np.array([
    [32, 33, 31, 29, 35],   # Cidade A : temp1, temp2, temp3, temp4, temp5
    [28, 27, 30, 26, 29],   # Cidade B
    [22, 24, 21, 23, 25]    # Cidade C
])

# A temperatura da Cidade B no 3º dia (índice 2)
print(temperaturas[1, 2])
print()

# Todas as temperaturas da Cidade C
print(temperaturas[2, :])
print()
# As temperaturas de todas as cidades no último dia
print(temperaturas[:, -1])
print()
# A submatriz das cidades A e B nos primeiros 3 dias
submatriz_temp = temperaturas[0:2, 0:3]
print(submatriz_temp)
````

### Exercício 5 — Desafio: Mini Análise de Dataset com Matrizes

**Enunciado:** Uma empresa de e-commerce tem os dados de 5 produtos em 4 regiões do Brasil (Norte, Nordeste, Sudeste, Sul), representando as **quantidades vendidas** no mês:

```python
quantidades = np.array([
    [120, 340, 890, 210],   # Produto 1
    [ 80, 150, 560, 300],   # Produto 2
    [200, 420, 1100, 480],  # Produto 3
    [ 50,  90,  320, 180],  # Produto 4
    [310, 500, 1450, 620]   # Produto 5
])

precos = np.array([29.90, 49.90, 19.90, 89.90, 14.90])  # Preço de cada produto
```

Resolva **sem usar laços `for`**:

1. Calcule a receita total gerada por cada produto em todas as regiões (soma das colunas de cada linha × preço).
2. ''Calcule o total de unidades vendidas por região (soma de cada coluna).
3. Identifique qual produto e qual região têm o maior e o menor volume de vendas.
4. Crie uma matrix de receitas `(5, 4)` onde cada célula é `quantidade × preço_do_produto` para aquela região. *(Dica: use reshape ou broadcasting 2D.)*
5. Exiba a receita total por região.''

> 💡 **Dica para o item 4:** Tente `quantidades * precos.reshape(5, 1)` — isso multiplica cada linha pelo preço correspondente daquele produto (broadcasting 2D).


````python
import numpy as np

quantidades = np.array([
    [120, 340, 890, 210],   # Produto 1
    [ 80, 150, 560, 300],   # Produto 2
    [200, 420, 1100, 480],  # Produto 3
    [ 50,  90,  320, 180],  # Produto 4
    [310, 500, 1450, 620]   # Produto 5
])

precos = np.array([29.90, 49.90, 19.90, 89.90, 14.90])  # Preço de cada produto

# Calcule a receita total gerada por cada produto em todas as regiões (soma das colunas de cada linha x preço)
receita_total = np.sum(quantidades, axis=1) * precos
print("Receita total por produto:", receita_total)
print()

# 2. Calcule o total de unidades vendidas por região (soma de cada coluna).
total_por_regiao = np.sum(quantidades, axis=0)
print("Total de unidades vendidas por região:", total_por_regiao)
print()

# 3. Identifique qual produto e qual região têm o maior e o menor volume de vendas.
produto_mais_vendido = np.argmax(np.sum(quantidades, axis=1))
produto_menos_vendido = np.argmin(np.sum(quantidades, axis=1))
regiao_mais_vendida = np.argmax(np.sum(quantidades, axis=0))
regiao_menos_vendida = np.argmin(np.sum(quantidades, axis=0))
print(f"Produto mais vendido: Produto {produto_mais_vendido + 1}")
print(f"Produto menos vendido: Produto {produto_menos_vendido + 1}")
print(f"Região mais vendida: Região {regiao_mais_vendida + 1}")
print(f"Região menos vendida: Região {regiao_menos_vendida + 1}")
print()

# 4. Crie uma matrix de receitas `(5, 4)` onde cada célula é `quantidade × preço_do_produto` para aquela região. 
# #*(Dica: use reshape ou broadcasting 2D.)*
precos = precos.reshape(5,1)
receitas = quantidades * precos
print("Matriz de receitas por produto e região:")
print(receitas)
print()

# 5. Exiba a receita total por região.
receita_por_regiao = np.sum(receitas, axis=0)
print("Receita total por região:", receita_por_regiao)  
````
