## 📖 Bloco Teórico + Analogias

### O que é o Pandas?

**Pandas** (Python Data Analysis Library) é a biblioteca mais usada no mundo para manipulação e análise de dados tabulares em Python. Foi criada em 2008 e é construída sobre o NumPy — ou seja, por baixo dos panos, um DataFrame do Pandas é uma coleção organizada de arrays NumPy.

Ela oferece duas estruturas principais de dados: **Series** (unidimensional) e **DataFrame** (bidimensional). Com o Pandas você consegue:

- Ler e salvar dados em dezenas de formatos (CSV, Excel, Parquet, JSON, SQL, etc.)
- Selecionar, filtrar e transformar colunas com sintaxe intuitiva
- Agrupar, pivotar e cruzar tabelas
- Tratar valores ausentes e duplicados
- Fazer análise exploratória completa em poucas linhas

> 🏦 **Analogia do extrato bancário:** O NumPy é como ter só os números da sua conta em uma lista — você sabe que existem, mas não sabe o que cada um significa. O Pandas é como o extrato bancário completo: cada linha tem data, descrição, valor e saldo. Os dados têm **contexto**.

---

### Series: a coluna inteligente

Uma **Series** é uma estrutura unidimensional — como uma única coluna de planilha — com duas partes: os **valores** (os dados em si) e o **índice** (o rótulo de cada valor). Por padrão o índice é 0, 1, 2, 3…, mas você pode usar qualquer rótulo: datas, nomes, códigos.

> 📋 **Analogia da lista de chamada:** Em uma lista de chamada escolar, cada aluno tem um número (índice) e um nome (valor). Se você usar o nome como índice, pode chamar diretamente por nome em vez de número. Isso é o índice personalizado do Pandas.

A Series aceita qualquer tipo de dado: inteiros, decimais, texto, booleanos, datas.

### DataFrame: a tabela inteligente

Um **DataFrame** é uma estrutura bidimensional — linhas e colunas, como uma planilha Excel ou uma tabela de banco de dados. Você pode pensar nele como um **dicionário de Series**: cada coluna é uma Series, e todas compartilham o mesmo índice (as linhas).

> 📊 **Analogia da planilha do mercado:** Imagine a planilha de controle de estoque de um mercado. Cada linha é um produto. Cada coluna é um atributo: nome, categoria, preço, quantidade. O DataFrame é essa planilha — mas com superpoderes.

A relação entre Series e DataFrame é simples:
- Um DataFrame com **uma coluna** é como uma Series com nome.
- Um DataFrame é **várias Series lado a lado**, todas com o mesmo índice de linhas.

### Leitura de Dados: pd.read_csv()

`pd.read_csv()` é provavelmente a função mais usada em projetos de Data Science. Ela lê arquivos de texto com valores separados por um delimitador (vírgula, ponto-e-vírgula, tabulação) e retorna um DataFrame.

Os parâmetros mais importantes:

| Parâmetro | Para que serve | Exemplo |
|-----------|---------------|---------|
| `sep` | Define o separador entre colunas | `sep=';'` para arquivos brasileiros |
| `encoding` | Define a codificação de caracteres | `encoding='latin1'` para acentos |
| `header` | Número da linha que contém o cabeçalho | `header=0` (padrão) |
| `index_col` | Coluna a usar como índice | `index_col='id'` |
| `nrows` | Lê apenas as primeiras N linhas | `nrows=100` para testar |

> 🗺️ **Analogia do mapa de tesouro:** O arquivo CSV é o mapa — cheio de informações, mas num formato que você precisa saber decifrar. `pd.read_csv()` é o tradutor: você diz qual é o separador, qual é o encoding, e ele transforma aquele arquivo de texto numa tabela organizada.

**Atenção aos arquivos brasileiros:** No Brasil, é muito comum que arquivos CSV usem `;` como separador (não `,`) porque a vírgula é usada como separador decimal (R$ 1.234,56). Sempre verifique antes de ler!

---

### Leitura de Dados: pd.read_excel() e pd.read_parquet()

**`pd.read_excel()`** lê arquivos `.xlsx` e `.xls`. Parâmetros importantes: `sheet_name` (qual aba ler) e `header` (linha do cabeçalho).

**`pd.read_parquet()`** lê arquivos `.parquet` — um formato binário colunar muito eficiente, cada vez mais comum em pipelines de dados. É muito mais rápido e compacto que CSV para datasets grandes.

> 💾 **Analogia dos formatos:** CSV é como um caderno escrito à mão — qualquer um abre, mas demora para escrever e ler. Excel é como um caderno impresso com formatação — bonito, mas pesado. Parquet é como um HD externo comprimido — menor, mais rápido, mas precisa de um leitor específico.

### Gravação de Dados

Depois de processar os dados, você precisa salvar o resultado. Os métodos de gravação espelham os de leitura:

| Método | Salva como |
|--------|-----------|
| `df.to_csv('arquivo.csv')` | Arquivo CSV |
| `df.to_excel('arquivo.xlsx')` | Planilha Excel |
| `df.to_parquet('arquivo.parquet')` | Arquivo Parquet |

O parâmetro `index=False` é quase sempre necessário em `to_csv()` e `to_excel()` — sem ele, o Pandas salva o índice (0, 1, 2…) como uma coluna extra no arquivo.

**Erro 4: Problemas de encoding ao ler arquivos com acentos**

```python
# ❌ PROBLEMA: arquivo salvo em latin1 lido como utf-8 (ou vice-versa)
# UnicodeDecodeError: 'utf-8' codec can't decode byte ...

# ✅ SOLUÇÃO: tentar os encodings mais comuns para arquivos brasileiros
# Ordem de tentativas:
# 1. encoding='utf-8'       → padrão moderno
# 2. encoding='utf-8-sig'   → utf-8 com BOM (arquivos exportados do Excel)
# 3. encoding='latin1'      → ISO-8859-1, muito comum em sistemas legados
# 4. encoding='cp1252'      → Windows-1252, variante do latin1

try:
    df = pd.read_csv('arquivo.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('arquivo.csv', encoding='latin1')
    print("Arquivo lido com encoding latin1")
```
